```python
import graph_tool.all as gt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read the corruption network
g = gt.load_graph("dataset/test_review/corruption.gt")

# Get basic network statistics
print(f"Number of nodes: {g.num_vertices()}")
print(f"Number of edges: {g.num_edges()}")

# Create a nice visualization
pos = gt.sfdp_layout(g) # Force-directed layout
gt.graph_draw(g, pos=pos, 
              vertex_size=5,
              edge_color="lightgray",
              output_size=(1200,1200),
              output="corruption_network.png")

plt.figure(figsize=(8,8))
img = plt.imread("corruption_network.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import networkx as nx
from pathlib import Path

# Read the Digg network using networkx
digg_path = "dataset/test_review/Digg.txt"
G = nx.read_edgelist(digg_path)

# Convert to graph-tool graph
gt_graph = gt.Graph(directed=False)
gt_graph.add_vertex(n=len(G.nodes()))

# Add edges
edge_list = list(G.edges())
for e in edge_list:
    gt_graph.add_edge(int(e[0]), int(e[1]))

# Save in same directory as input
save_path = str(Path(digg_path).with_suffix('.gt'))
print(save_path)
gt_graph.save(save_path)

print(f"Converted network saved to {save_path}")
print(f"Number of nodes: {gt_graph.num_vertices()}")
print(f"Number of edges: {gt_graph.num_edges()}")
```
